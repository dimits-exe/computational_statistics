a) We will start by creating 50 different data sets with 30 elements each (aka 50 numpy arrays of size 30)

In order to make the process simple, we will build a function that returns instances of the Dataset. Note how we include a 1-like vector in our x data; since we investigate the linear assumption, we must assume a coefficient may exist.

```python
import numpy as np
import math

# generator instance to centralize all RNG calls to
gen = np.random.default_rng()

def generate_dataset(size: int, theta_0: float) -> tuple[np.ndarray, np.ndarray]:
    """
    Create a y=theta_+0*x (x in unif [0, 100]) dataset of N elements, with Gaussian noise, adding ones to the x matrix as 
    to support finding coefficients under the linear assumption.
    @param size: the N elements of the dataset
    @param theta_0: the linear coefficient
    return: a tuple containing a N-sized vector (y) and a Nx2 matrix (x)
    """
    unif_low = 0
    unif_up = 100 
    noise_mean = 0
    noise_var = 64
        
    x_orig = gen.uniform(unif_low, unif_up, size)
    y = theta_0 *x_orig
    # insert noise into our "perfect" dataset 
    # the "normal" method uses stddev instead of variance
    x =  x_orig + gen.normal(noise_mean , math.sqrt(noise_var), size) 
    X = np.column_stack([np.ones_like(x), x])
    return y, X
```

```python
# we use a python list instead of a NxM matrix of datasets for the sake of code clarity
datasets = [generate_dataset(30, 2) for i in range(50)]
datasets[0][0].shape, datasets[0][1].shape
```

b) We will now import the Ordinary Least Squares and Mean Squared Error functions we defined in the previous notebook. 

As we will see later, we will also need the Ridge Regression algorithm as well as our original OLS. Since Ridge is a modification (parameter constraint) of OLS, we can easily modify the function to support it. 

```python
def least_squares(X: np.array, y: np.array, lamda: float = 0) -> np.array:
    """
    Perform linear regression using the sum of error squares criterion.
    @param X: a NxM matrix containing the data
    @param y: a Nx1 vector containing the observed values of the data
    @param lamda: the ridge regression term, 0 for ordinary least squares (default 0)
    @return: the theta vector of size Mx1 containing the model parameters
    """
    if lamda < 0:
        raise ValueError("Lamda values must be equal to, or larger than 0.")
    
    # if X is a vector and not matrix, turn it into a matrix as
    # vector @ vector => float, but matrix @ vector => vector
    if X.ndim < 2:
        X = X.reshape(-1, 1)
    
    XTX = X.T@X
    inverted = XTX + lamda * np.identity(XTX.shape[0]) # ridge regression term

    if np.linalg.det(inverted) == 0:
        print("Warning: X^TX not invertible, solution is not unique")
    
    # mutiply the inverse of X^T with X^T and y
    # the result is our theta vector
    return np.linalg.inv(inverted) @ X.T @ y 
```

```python
def mse(y_hat: np.array, y: np.array) -> float:
    """
    Calculate the Mean Squared Error between the predictions and actual values.
    @param y_hat: A Nx1 prediction vector
    @param y: A Nx1 vector containing the actual values
    @return: the MSE error
    """
    return np.mean((y - y_hat)**2)
```

Now that we have defined our algorithms, let's run them on all 50 datasets and store the parameter results.

```python
# only grab the linear coefficient
thetas = np.array([least_squares(d[1], d[0]) for d in datasets])
thetas.shape
```

c) As always, we can compute our estimates for each dataset by multiplying the data and parameter vectors. 

```python
# since theta is scalar, we can just multiply instead of using matrix multiplication
y_hat_ls = [x[1] @ theta for x, theta in zip(datasets, thetas)]
```

We can compute the MSE between the random variable ($\hat{\theta}$) and the real model $(\theta_0=2)$ as follows:

```python
# only compare the linear coefficient with theta_0
theta_0_hat = thetas[:,1]
theta_0 = np.full_like(theta_0_hat, 2)
mse(theta_0_hat, theta_0)
```

```python
import matplotlib.pyplot as plt

plt.plot(theta_0_hat)
plt.plot(theta_0, color="red")

plt.title("Theta Estimator Variance")
plt.xlabel("Dataset used")
plt.ylabel("Theta parameter estimation")

plt.legend(["Estimated theta", "Ground Truth (Theta 0)"])

plt.show()
```

In any case, the fact that our estimations are hovering around the global minimum (aka the real distribution), tells us that the estimator, that being the parametric set $y = ax + b$, is almost correct.

b) In order to create an 8th degree polynomial LS estimator we must transform our original 10 points into a 10x9 matrix (we add one column of 1s for the coefficient).

We thus create a new dataset where each column is the original vector raised to the power of that column's index (so $x^0, x^1, \cdots, x^8$).

```python
def polynomial(x: np.ndarray, degree: int) -> np.ndarray:
    """
    Raise a vector to a N-degree polynomial.
    @param x: a Lx1 sized vector
    @param degree: the degree of the polynomial
    @return: a Lx(degree+1) vector representing the Nth degree polynomial of the original vector 
    """
    x8 = np.column_stack([np.power(x, i) for i in range(degree + 1)])
    
    # discard useless 3rd dimension if exists
    if x8.ndim == 3: 
        return x8[:,:,0]
    else:
        return x8
        

X8 = polynomial(X, 8)
X8.shape
```

And get the estimated parameters for the new dataset by using our now-familiar LS function.

```python
theta = least_squares(X8, y)
theta
```

We can visually compare the efficiency of the LS estimation by plotting the estimations against the original data. 

Note: In order to plot the generated polynomial functions, we will use 100 evenly-space points instead of our original dataset in order ot get a smooth curve. The model parameters of course will be generated by the original (transformed) dataset.

```python
# 100 evenly spaced points
x_plot = np.linspace(0, 1, 100)
# our samples transformed to the 8-dimensional space
x_plot_poly = polynomial(x_plot, 8)

x_plot.shape
```

```python
# our estimations for the 100 points
y_plot = x_plot_poly @ theta
y_plot.shape
```

```python
plt.scatter(X, y, color="red", marker="*", label="Ground Truth")
plt.plot(x_plot, y_plot, color="blue", label="8th degree polynomial")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Ground Truth vs Polynomial Estimation")
plt.legend()

plt.show()
```

Here we can see a peculiar model, which tightly "hugs" the data but drastically (and to us, illogically) curves towards the end, where the data points end. 

This model will obviously have a very low MSE, since it technically covers all data points, but fails to acknowledge their linear nature. We will discuss the reprecussions of this approach at the end of the exercise.

We can attempt to plot a representative sample of our 100-hyperparameter long array, to investigate the reasons for the clear upward MSE trend for larger $\lambda$ values:

```python
plot_lamdas = np.insert(lamdas[0::30], 0 , 0)
plot_lamdas
```

```python
import matplotlib

def plot_ridges(axis: matplotlib.axes._axes.Axes, lambdas: np.ndarray) -> None:
    # plot ground truth
    axis.scatter(X, y, color="blue", marker="*", label="Ground Truth")
    
    # plot curves
    for plot_lamda in lambdas:
        estimate = x_plot_poly @ least_squares(X8, y, lamda=plot_lamda)
        axis.plot(x_plot, estimate, label=f"Lambda={plot_lamda:.3f}")
        
    axis.legend()
    axis.set_xlabel("x")
    axis.set_ylabel("y")        
        
        
fig, (ax1, ax2) = plt.subplots(1,2)
plot_ridges(ax1, plot_lamdas[:3])
plot_ridges(ax2, plot_lamdas)
ax1.set_title("Small lambda values")
ax2.set_title("All lambda values")
fig.tight_layout(pad=5.0)
fig.set_figheight(6)
fig.set_figwidth(11)
fig.suptitle("Ridge Regression: Lambda Parameter Impact on Least Squares")

plt.show()
```

We have split the plot in two subplots in order to effectively show the differences in the curves of the smaller lambda values.

d) At this point we have enough data to make some observations.

- First of all, it's evident that we have chosen the wrong parametric set for our estimator. We have a 10 data-point long dataset and are trying to use a very complex model with 9 (!) parameters. Our original OLS estimator can cope by "copying" the data points, hence the extremely large parameter values. By imposing restrictions on those parameters however (by using larger lambda values in Ridge), this flexibility is forcefully removed from it, and thus its efficiency suffers.

- Sufficiently large values of lambda result in a complete drift away from the original distribution. This is caused by inserting too much bias in the model, in this case forcing its parameters to be extremely close to 0.

- Lower lambda values result in much less drastic changes in our curve, leading it to adopt something more akin to a straight line. While this results in lower MSE scores, it leads to a much more understandable model.

What we have essentially showcased is the effect of regularization on overfitting. Our original model is too powerful (too many parameters) for our small, easy to model dataset, and thus encodes the data points and their noise perfectly in its parameters. This comes at the cost of being extremely variant (it would likely completely change if we decided to add more data points) and hard to use for explaining the underlying distribution. 

Our regularized model on the other hand, is technically worse on MSE scores, but does not suffer from the problems outlined above.
