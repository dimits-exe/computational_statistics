### Exercise 3:

Consider the exclusive OR (XOR) classification problem, where the points (0,0) and (1,1) are assigned to class ‚àí1 and the points (0,1) and (1,0) are assigned to class +1.

##### 1. Draw the points on the paper and prove that the classification problem is not linearly separable.

#### Answer

We can visually confirm that the classes aren't linearly separable by plotting them:

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot positive and negative points with different colors
x = np.array([0, 1, 0, 1])
y = np.array([1, 0, 0, 1])
labels = np.array([1,1,-1,-1])

scatter = plt.scatter(x, y, c = labels)

plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(*scatter.legend_elements(), loc="center right")

# Create a title
plt.title("XOR")

plt.show()
```

As we can see, there is no line which can separate all the points correctly.

We can prove this mathematically:


Assume that the problem is linearly separable. Then, there exists a function $f(x_1, x_2) = \theta_0 + \theta_1 x_1 + \theta_2 x_2 $ where:
$f(x_1, x_2) > 0 \text{ for } x_0 = 1, x_2 = 0, \text{ or } x_1 = 0, x_2 = 1$

and 

$f(x_1, x_2) < 0 \text{ for } x_1 = 0, x_2 = 0, \text{ or } x_1 = 1, x_2 = 1$ 


Subsituting $x_1$ and $x_2$ we get:

$f(1,0) > 0 \iff  \theta_0 + \theta_1 > 0$ (1)

$f(0,1) > 0 \iff  \theta_0 + \theta_2 > 0$ (2)

$f(0,0) < 0 \iff  \theta_0 < 0$ (3)

$f(1,1) < 0 \iff  \theta_0 + \theta_1 + \theta_2 < 0 $ (4)

From (1), (3) we can infer that $\theta_1 > 0$ (5)

From (2), (3) we can infer that $\theta_2 > 0$ (6)

From (4), (3) we can infer that $\theta_0 > \theta_1 + \theta_2$

which is impossible, since from (3) we established that $\theta_0 < 0$, and from (5), (6) that $\theta_1 > 0$ and $\theta_2 > 0$ respectively.

Thus the XOR problem is NOT linearly separable.

##### 1. Use $\theta_0 = 3, \theta_1 =2, \theta_2 = 1, \theta_3 = 1 $(adopt the strategy given in the example of the 1st slide of the 2nd lecture). In the sequel, pretend that you do not know the model that generates the data. All you have at your disposal is the data set $ùëã$.

**Answer:**

We define our data vectors as:

```python
# Generate data points from model
SIZE = 200
x1 = np.random.uniform(0, 10, SIZE)
x2 = np.random.uniform(0, 10, SIZE)
eta = np.random.normal(0, 0.05)

y = 3 + 2*x1 + x2 + x1*x2 + eta
```

###### 2. Adopting the linear model assumption in the original space (that is, assuming that $ùë¶ = \theta_0 + \theta_1ùë•_1 + \theta_2ùë•_2$), estimate the parameters of the model $(\theta_0, \theta_1, \theta_2)$ that minimize the sum of error squares criterion.

**Answer:**

We will create a data matrix $X$ where the first column is composed of `1`s (in order to calculate the constant $\theta_0$) and the other columns the $x_1$ and $x_2$ vectors.

```python
X = np.array([np.ones(shape=(SIZE)), x1, x2]).T

print(X[1:5])
X.shape
```

We will now write the OLS algorithm, using the sum of error squares criterion. We will use `numpy`'s numerical operations for the sake of speed, clarity and code reliability. 

We will define the algorithm as a function in order to use it without duplicating code.

```python
def ols(X: np.array, y: np.array) -> np.array:
    """
    Perform linear regression using the sum of error squares criterion.
    @param X: a NxM matrix containing the data
    @param y: a Nx1 vector containing the observed values of the data
    @return: the theta vector of size Mx1 containing the model parameters
    """
    XTX = X.T@X # multiply X^T with X

    if np.linalg.det(X.T@X) == 0:
        print("Warning: X^TX not invertible, solution is not unique")
    
    # mutiply the inverse of X^T with X^T and y
    # the result is our theta vector
    return np.linalg.inv(XTX) @ X.T @ y 


theta = ols(X, y)
```

We can also make a quick function to format our theta vector, as we will be checking it frequently in this notebook.

```python
def repr_theta(theta: np.array) -> str:
    """
    Get the str representation of a theta (parameter) vector.
    @param theta: the theta vector
    @return: a string describing the vector
    """
    s = ""
    for i in range(theta.shape[0]):
        s += f"Theta_{i} = {theta[i]}, "  
    return s
```

```python
print(repr_theta(theta))
```

##### 3. For each one of the 200 data points $x_i$ of ùëã, determine the associated estimate $\hat {y_i}$ provided from the model estimated in (2) and compute the $ùëÄùëÜùê∏ = \frac{1}{200}\sum^{200}_{i=1}(y_i-\hat{y_i})^2$.

**Answer:**

We can obtain the $\hat{y_i}$ vector by multiplying our parameter vector with our data matrix.

```python
y_hat = X @ theta
```

The MSE can be defined in `numpy` as follows:

```python
def mse(y_hat: np.array, y: np.array) -> float:
    """
    Calculate the Mean Squared Error between the predictions and actual values.
    @param y_hat: A Nx1 prediction vector
    @param y: A Nx1 vector containing the actual values
    @return: the MSE error
    """
    return 1 / y_hat.shape[0] * np.sum(np.power(y - y_hat, 2))

print("Mean Squared Error for the linear assumption model: ", mse(y_hat, y))
```

##### 7. Comment on the results obtained in (3) and (6).

**Answer:**

In the case of the linear assumption, we note a relatively high MSE, meaning that our model cannot sufficiently fit the data. This is caused by our choise of projection; our model attempts to fit quadratic data (there is an interaction between the $x_1, x_2$ terms) in a linear space, which leads to our model "losing" valuable information. This is also the reason the estimated parameters are way off compared to our generator function's.

In the case of a quadratic assumption, where we "guessed" correctly the generator function's parametric set (the *kind* of parameters used, not their values), we note that the MSE is practically 0, and the estimated parameters are almost exactly correct. Thus our model can almost perfectly predict the data.

We can conclude therefore that estimating the correct parametric set of the generator function is essential to correctly and accurately estimating the distribution of the data points. Even an accurate and mathematically sound method will fail if we fail to identify the underlying distribution of our data.
